데이터 병합

```python
import pandas as pd

for i in range(300):
    if i == 0:
        info = pd.read_csv(f'data/place_info_{i+1}.csv')
        review = pd.read_csv(f'data/place_review_{i+1}.csv')
    else:
        try:
            info_new = pd.read_csv(f'data/place_info_{i+1}.csv')
            info = pd.concat([info, info_new], ignore_index=True) # 이전데이터에 인덱스를 가져온다 해결방안 ignore_index=True
        except pd.errors.EmptyDataError:
            None
        try:
            review_new = pd.read_csv(f'data/place_review_{i+1}.csv')
            review = pd.concat([review, review_new], ignore_index=True)
        except pd.errors.EmptyDataError:
            None
```

```python
# place_name null 값 제거 
info.dropna(subset=['place_name'], inplace=True)
```

```python
# 중복 제거 
info=info.drop_duplicates(subset='place_id', keep='first')
```

```python
info_review=pd.merge(info, review, on='place_id', how='left')
info_review
```

전처리

```python
info_review.info()
```

```python
# tag_list
info_review['tag_list'] = info_review['tag_list'].fillna(" ")
```

```python
# head_image
info_review.drop('head_image', axis=1, inplace=True)
```

```python
# user_name
nan_indices = info_review[info_review['user_name'].isnull()].index
info_review.loc[nan_indices, 'user_name'] = ['anonymous' + str(i) for i in range(1, len(nan_indices) + 1)]
```

```python
# content
info_review['content'] = info_review['content'].fillna(" ")
```

```python
# user_rating
# user_rating이 NaN인 경우 place_id로 그룹화하여 comm_score로 대체

# place_id로 그룹화하여 각 그룹의 comm_score를 추가
info_review['comm_score_group'] = info_review.groupby('place_id')['comm_score'].transform('first')

# user_rating이 NaN인 경우 comm_score_group으로 대체
info_review['user_rating'] = info_review['user_rating'].fillna(info_review['comm_score_group'])

# comm_score_group 컬럼 삭제 (필요에 따라)
info_review.drop(columns=['comm_score_group'], inplace=True)
```

```python
# create_time
info_review['create_time'] = info_review['create_time'].fillna("2018-01-01 00:00:00")
```
