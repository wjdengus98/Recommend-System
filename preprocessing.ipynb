```python
import pandas as pd
info= pd.read_csv('data/place_info_105.csv')
info

review= pd.read_csv('data/place_review_105.csv')
review
```

```python
info_review=pd.merge(info, review, on='place_id', how='left')
info_review
```

```python
info_review.info()
```

```python
info_review['review_count']= info_review.groupby('place_id')['content'].transform('count')
info_review
```

```python
info_review_ex1=info_review.copy()
info_review_ex2=info_review.copy()
```

```python
info_review_ex_1 =info_review_ex1[info_review_ex1['review_count']>=5].reset_index(drop=True)
info_review_ex_2 =info_review_ex2[info_review_ex2['review_count']< 5].reset_index(drop=True)
```
