데이터 병합

```python
import pandas as pd

for i in range(300):
    if i == 0:
        info = pd.read_csv(f'data/place_info_{i+1}.csv')
        review = pd.read_csv(f'data/place_review_{i+1}.csv')
    else:
        try:
            info_new = pd.read_csv(f'data/place_info_{i+1}.csv')
            info = pd.concat([info, info_new], ignore_index=True) # 이전데이터에 인덱스를 가져온다 해결방안 ignore_index=True
        except pd.errors.EmptyDataError:
            None
        try:
            review_new = pd.read_csv(f'data/place_review_{i+1}.csv')
            review = pd.concat([review, review_new], ignore_index=True)
        except pd.errors.EmptyDataError:
            None
```

```python
info_review=pd.merge(info, review, on='place_id', how='left')
info_review
```

전처리 임의 값 추가

```python
info_review_ex=info_review.copy()
```

```python
# tag_list에서 null 값은 -> 상관없음으로 변경
info_review_ex['tag_list'] = info_review_ex['tag_list'].fillna("It doesn't matter")
```

```python
info_review_ex.info()
```

```python
info_review_ex.dropna(subset=['place_name'], inplace=True)
```

```python
info_review_ex.info()
```

```python
# 'content' 컬럼의 결측값을 빈 문자열로 채우기
info_review_ex['content'] = info_review_ex['content'].fillna('')
```

```python
# 'head_image' 컬럼의 결측값을 'unknown_image'로 채우기
info_review_ex['head_image'] = info_review_ex['head_image'].fillna('unknown_image')
```

```python
# 'user_name' 컬럼의 결측값을 'anonymous'로 채우기
info_review_ex['user_name'] = info_review_ex['user_name'].fillna('anonymous')
```

```python
info_review_ex.info()
```

```python
info_review_ex['user_rating']=info_review_ex['user_rating'].fillna(0)
```

```python
info_review_ex.info()
```

```python
info_review_ex['create_time']=info_review_ex['create_time'].fillna("0000-00-00 00:00:00")
```

```python
info_review_ex.info()
```

```python
info_review_ex.to_csv('info_review_ex.csv', index=False)
```
